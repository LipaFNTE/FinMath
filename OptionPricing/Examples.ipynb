```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from random import gauss
```

Będziemy wyceniać opcję europejską call zadaną wzorem $C_T = (S_T - K)^+$. W tym przypadku zarabiamy za każdym razem gdy w czasie T cena naszego instrumentu będzie większa niż K. Musimy jeszcze zdyskontować naszą wypłatę (czyli trochę obniżyć jej wartość, bo w tym samym czasie moglibyśmy bez ryzyka wrzucić całą kasę z inwestycji na lokatę). Z tego względu finalna cena to $C_T = \exp(- r * T) * (S_T - K)^+$

```python
sigma = 0.2
r = 0.05
S_0 = 98
K = 100
T = 1
n_min = 100
n_max = 50000
step = 100
```

Zgodnie z metodą Monte Carlo możemy wygenerować cenę $S_T$ zgodnie z postulatami Blacka-Scholesa. Wtedy cena jest równa $S_T = S_0 \exp ((r - 0.5 * \sigma^2) T + \sigma \sqrt(T) Z$ gdzie Z jest zmienną losową o standardowym rozkładzie normalnym. Następne liczymy wartość $C_T$ i powtarzamy czynność n razy. Zgodnie z prawem wielkich liczb ciąg ten wraz ze wzrostem n będzie zbiegał z prawdopodbieństwem 1 do prawdziwego $C_T = E[\exp(-r T) (S_T - K)^+]$

```python
def generate_asset_price(S,v,r,T):
    return S * np.exp((r - 0.5 * v**2) * T + v * np.sqrt(T) * gauss(0,1.0))

def call_payoff(S_T,K):
    return max(0.0,S_T-K)
```

Teraz n-krotnie wykonujemy symulację dla różnych n. Prawdziwa cena według kalkulatora (akurat to jest prosta opcja, więc mamy tutaj sytuację, gdzie mamy analityczne rozwiązanie) jest równa 9.22

```python
p_offs = {}
for i in range(n_min, n_max, step):
    p_offs[i] = np.exp(-r * T) * np.mean([call_payoff(generate_asset_price(S_0, sigma, r, T), K) for _ in range(i)])
```

```python
plt.plot(list(p_offs.keys()), list(p_offs.values()))
```

Teraz zrobimy plot MSE

```python
MSE_option = [(k - 9.22)**2 for k in list(p_offs.values())]
```

```python
plt.plot(list(p_offs.keys()), MSE_option)
```

Tak jak wspomniałem, tutaj sytuacja jest prosta, bo mamy prostą opcję i da się tę wartość oczekiwaną policzyć normalnie zgodnie z probabilistycznymi metodami. Niestety są przypadki, gdzie jest już to dużo trudniejsze lub nawet niemożliwe (szczególnie w większych wymiarach dla opcji zależnych nie od jednego $S_T$ ale od kilku aktywów lub w przypadku gdy $S_T$ jest zdefiniowane dużo bardziej skomplikowanie). Przykłady takich opcji to opcje górskie (Himalajska, Altiplano, Exchange)
