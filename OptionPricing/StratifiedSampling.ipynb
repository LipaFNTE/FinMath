```python
import Stratified_Sampling, Payoff
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Zadanie 1

Korzystam z klas Stratified_Sampling oraz Payoff, w których zdefiniowałem metody związane z metodami Monte Carlo (stratified sampling) oraz z wypłatami z opcji (w tym przypadku call)

W pierwszej kolejności podam wynik mojego rozwiązania dla przykładowej opcji barierowej. Wybieram poszczególne parametry i tworzę obiekt klasy StratifiedSampling, który zawiera wszystkie potrzebne metody do przedstawienia wyniku. Tworzę obiekt klasy BarrierOut, który dziedziczy po klasie TrajectoryPayoff, która dziedziczy po Payoff.

```python
A, B = 82, 125
call = True
```

```python
n = 50000
k = 50
T = 1
r = 0.001
mu = 0
sigma = 0.2
s_0 = 100
K = 105
M = 10
```

```python
ss = Stratified_Sampling.StratifiedSampling(n, k, T, r, mu, sigma, s_0, K, M)
payoff = Payoff.BarrierOut(A, B)
res = ss.run(True, payoff, call)
```

Wypisuję wynik, który utworzył się w obiekcie klasy Result, aby móc mieć wszystko w jednym obiekcie.

```python
res.print_result()
```

Teraz przedstawię statystyki odnośnie ceny i RMSE w zależności od liczby wartstw. Dane są takie same, jak w poprzednim punkcie, zmieniam n na 100000 oraz manipuluję k.

```python
n = 100000
k = [10, 20, 25, 50, 100, 200, 500, 1000, 2000, 2500, 5000]
res = []
for i in k:
    ss = Stratified_Sampling.StratifiedSampling(n, i, T, r, mu, sigma, s_0, K, M)
    payoff = Payoff.BarrierOut(A, B)
    res.append(ss.run(True, payoff, call))
```

Wykres ceny w zależności od parametru k

```python
prices = []
rmse = []
for result in res:
    if isinstance(result.price, np.ndarray):
        prices.append(result.price[0])
    else:
        prices.append(result.price)
        
for result in res:
    if isinstance(result.rmse, np.ndarray):
        rmse.append(result.rmse[0])
    else:
        rmse.append(result.rmse)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(k, prices)
plt.title('Wykres ceny zależnej od liczby warstw')
plt.xlabel('Liczba warstw')
plt.ylabel('Cena')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(k, rmse)
plt.title('Wykres RMSE zależnej od liczby warstw')
plt.xlabel('Liczba warstw')
plt.ylabel('RMSE')
```

```python
for result in res:
    result.print_result()
```

Wyniki stabilizują się dla większej liczby warstw, RMSE mimo wahań na początku, potem powoli spada wraz ze wzrostem liczby warstw. Dziwne wydają się zachowania ceny oraz RMSE dla małej liczby warstw. Warto by było się przyjrzeć temu i spróbować znaleźć powód takiego zachowania.

Teraz będę manipulował parametrem M - liczbą punktów monitorowania barier (A, B). 

```python
A = 0
B = 145
```

```python
k = 100
M = [2, 5, 10, 20, 25, 50, 75, 100, 250, 500]
res_out = []
for i in M:
    ss = Stratified_Sampling.StratifiedSampling(n, k, T, r, mu, sigma, s_0, K, i)
    payoff = Payoff.BarrierOut(A, B)
    res_out.append(ss.run(True, payoff, call))
```

```python
prices = []
for result in res_out:
    if isinstance(result.price, np.ndarray):
        prices.append(result.price[0])
    else:
        prices.append(result.price)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(M, prices)
plt.title('Wykres ceny zależnej od liczby punktów monitorowania')
plt.xlabel('Liczba punktów monitorowania')
plt.ylabel('Cena')
```

Wykres, nielicząc kolejnych rozbieżności w początkowej fazie, jest bardzo logiczny. Wraz ze zwiększeniem liczby punktów monitorowania, mamy więcej możliwości, że znajdziemy się poza barierami, więc częściej nasza wypłata wynosi 0, co implikuje zmniejszenie ceny
